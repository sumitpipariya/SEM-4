### 09) WAP create a Scatter Plot with several colors in Matplotlib?

```python
import random

x = [random.randint(1,100) for i in range(20)]
y = [random.randint(1,100) for i in range(20)]
z = [random.randint(1,100) for i in range(20)]
a = [random.randint(1,100) for i in range(20)]


plt.scatter(x,y,color = 'red')
plt.scatter(y,x,color = 'blue')
plt.scatter(z,y,x,color = 'green')
plt.scatter(a,y,x,color = 'black')

plt.show()
```

### 10) WAP to create a Box Plot.

```python
import matplotlib.pyplot as plt
import numpy as np


data = np.random.normal(100, 10, 200)

fig = plt.figure(figsize =(10, 7))

ax = fig.add_axes([0, 0, 1, 1])

bp = ax.boxplot(data)
plt.show()
```
